> s = '  first star   :   second star   :    third star  '

* 1) for ~ in 문을 이용한 코드를 작성하시오.
* 2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
s = '  first star   :   second star   :    third star  '

# for ~ in 문법
L = s.split(':')
for i, x in enumerate(L):
    L[i] = x.strip()
print L
# List comprehension
L = [x.strip() for x in s.split(':')]
print L
```

        * http://www.naver.com
        * http://www.daum.net
        * http://www.nytimes.com

```python
import urllib2
import string

def ParsingPage(ss):
    source = urllib2.urlopen(ss).read()

    #print source # 추후 숙제 답안 제출 때엔 왼쪽 라인은 주석처리할 것
    
    # 먼저 '<'를 기준으로 나누어 준다.
    l = source.split('<')
    
    # 그다음 각각의 나뉘어진 원소에 대해서 '>'를 기준으로 나누어 2개 이상으로 나누어 졌다면, 오른쪽에 있는 문자열을 저장하고 그렇지 않으면 삭제한다.
    for x in range(len(l)):
        tempList = l[x].split('>')
        if len(tempList) > 1:
            # 구두 문자를 제거하고 값을 추가한다.
            tempString = tempList[-1]
            for punc in tempList[-1]:
                if ( punc in string.punctuation ):
                    tempString = tempString.replace(punc, '')
            l[x] = tempString
        else :
            l[x] = ' '
    
    # 모든 리스트의 값들을 다시 이어 붙인다.
    source = ' '.join(l)
    
    # dictionery 정의
    d = {}
    for x in source.split():
        if d.has_key(x):
            d[x] += 1
        else:
            d[x] = 1

    # 출력한다.
    import json
    print json.dumps(d, ensure_ascii=False, indent=4)
#
ParsingPage('http://www.naver.com')
ParsingPage('http://www.daum.net')
ParsingPage('http://www.nytimes.com')

```
