### 2. 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오. 

- 다형성(Polymorphism) : 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능
- 장점
      1. 적은 코딩으로 다양한 객체들에게 유사한 작업을 수행시킬 수 있다.
      2. 코드의 가독성을 높혀준다.

```python
class Student:
    def C_Language(self):
        print ("C언어 어려워요")
class Freshman(Student):
    def C_Language(self):
        print ("1학년 :하나도 몰라요...")
class Sophomore():
    def C_Language(self):
        print ("2학년 :왠만큼은 해요~")
class Junior(Student):
    pass
class Senior(Student):
    def C_Language(self):
        print ("4학년 :다 까먹었어요")
#Student 클래스를 상속받은 Freshman, Sophomore, Junior, Senior 클래스, 같은 C_Language()함수를 호출해도 제각기 동작합니다.      
L = (Student(),Freshman(),Sophomore(),Junior(),Senior())
for each in L:
    each.C_Language()
```

- Student 클래스를 상속받은 Freshman, Sophomore, Junior, Senior 클래스, 같은 C_Language()함수를 호출해도 제각기 동작합니다.
- Junior 클래스의 경우 C_Language()함수가 정의되어 있지 않지만 부모 클래스의 C_Language()함수를 사용할 수 있습니다. 

### 3.다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

```python
class Counter:
    # 생성자에 count 값과 step 값을 인자로 받을 수 있게 했습니다. count만 입력해도 됩니다.
    def __init__(self,count,step =1):
        self.count = count
        self.step = step
    # print 함수와 함께 객체가 호출 되었을 때 수행됩니다. 포맷팅을 사용해서 출력형식을 지정했습니다.
    def __str__(self):
        return '[Count(step:%s)]%s'%(self.step,self.count)
    # step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 구현했습니다.
    def incr(self):
        self.count += self.step
    #인스턴스 객체를 직접 호출(call)할 수 있게 하고, 호출 되었을 때 incr() 메소드를 호출했습니다.    
    def __call__(self):
        self.incr()
    # + 연산이 주어졌을 때 수행 될 수 있도록 메소드를 정의했습니다.
    def __add__(self,value):
        self.count += value
    # - 연산이 주어졌을 때 수행 될 수 있도록 메소드를 정의했습니다.
    def __sub__(self,value):
        self.count -= value
    #  관계연산 (>, <, ==)이 수행 될 수 있도록 메소드를 정의했습니다.
    def __cmp__(self,value):
        return self.count-value

c = Counter(10)
c()
print (c)
c.incr()
print (c)
c+3
print (c)
c-5
print (c)
print (c > 9)
print (c < 11)
print (c == 10)
print (Counter(10))
```

### 4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
            
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print (s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print (t)
```

- 자 위에서부터 한 줄씩 살펴보자 첫 째줄을 보니까 MySet은 list를 상속받았네? 그러면 지금 부터 MySet은 list의 모든 기능을 가져와 사용할 수 있어.
- 그 다음 코드를 보면 생성자 함수인 __init__을 정의하고 인수로 l을 받고 있네? 그러면 이제 하나의 인수를 받는 객체에서는 더 이상 list의 __init__가 아닌 MySet의 __init__을 사용할거야. __init__의 내용을 보니까 l의 원소를 각각 자신에게 append 하고 있는데 append는 현재 클래스에서 정의되어있지 않지? 그렇다면 이건 list의 append함수란걸 알 수 있어. MySet객체는 인수 하나가 주어지면 그걸 스스로에 append하는구나~ 그리고 나서 eliminate_duplicate란 함수를 사용하는데 이 함수에 대해서 보자.
- 어? 그런데 위에 @staticmethod라고 쓰여있네? 이 구문은 사용자가 컴퓨터에게 전하는 주석이라고 생각하면 되는데 컴퓨터에게 '이거는 staticmethod에요~'라고 알려주는 거야. 그러면 컴퓨터는 아래에 함수를 수행할때 static의 성질을 고려해서 수행하겠지 안에 내용은 간단한게 __init__에서 받은 인수중에서 중복되는걸 제거하는 코드야
- 그럼이제 MySet의 마지막 함수인 __str__을 볼까? __str__함수는 str() 또는 print에 의해 객체가 호출되었을때 자동적으로 수행되는 특별한 함수야 보통 객체의 문자열 정보를 출력하도록 해놓지. 여기에서도 객체가 리스트형으로 가지고 있는 정보를 {}안에 넣어서 출력하는거네?
- 자 이제 클래스에 대한 분석을 마쳤으니까  그 아래에 코드를 보자 맨 처음에 있는 조건문은 무슨 의미일까? __name__은 이 함수가 호출된 공간의 스코프 이름을 반환하는데, 그 스코프가 __main__일 경우 아래의 코드를 수행할 수 있어. __main__이란 것은 현재 코드가 수행되는 공간이 모듈로서 import되어 수행되어 지는게 아닌 인터프리터에서 직접 실행되는 공간이라는 뜻이야 지금 이 공간은 import되어진게 아니니까 __main__이겠지? 
- 이제 아래 코드가 실행될거야 비슷한 문장의 반복이니까 처음 두문장만 보자 변수 s에 MySet객체를 할당하고 , 리스트 인수를 주었어. 이제 이 인수는 __init__을 통해 eliminate_duplicate함수를 거쳐 중복이 제거된 채로 객체를 저장에 s안에 남을거야. 그 다음에 print를 통해 s를 불렀으니까 __str__이 호출되겠지? 그러면 s안에 중복을 제거된 채로 남아있던 원소들이 __str__에서 짜여진 틀에 맞춰 출력될거야 자 이해안되는거 있니?

### 5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    # | 연산으로 두 집합의 합집합을 반환하는 메소드를 구현했습니다.
    def __or__(self,otherSet):
        for ele in otherSet:#otherSet에 있는게
            if ele not in self:#나한테 없으면
                self.append(ele)#추가
    # & 연산으로 두 집합의 합집합을 반환하는 메소드를 구현했습니다.
    def __and__(self,otherSet):
        s = []#변경될 리스트 값
        for ele in self:#나한테 있는게
            if ele in otherSet:#otherSet에 있으면
                s.append(ele)#리스트에 추가
        self[:] = []#나를 비우고
        for ele in s:
            self.append(ele)#새리스트의 값을 추가
    # - 연산으로 두 집합의 합집합을 반환하는 메소드를 구현했습니다.       
    def __sub__(self,otherSet):
        for ele in self:#나한테 있는게
            if ele in otherSet:#otherSet에도 있으면
                self.remove(ele)#삭제
        
            
            
if __name__ == "__main__":
    s1 = MySet([1, 2, 3])
    t1 = MySet([3, 4, 5])
    s1 | t1
    print (s1)
    s2 = MySet([1, 2, 3])
    t2 = MySet([3, 4, 5])
    s2 & t2
    print (s2)
    s3 = MySet([1, 2, 3])
    t3 = MySet([3, 4, 5])
    s3 - t3
    print (s3)
```

```python
import urllib2
import string
from stopwords import *

#사전 객체 끼리 +연산이 가능하도록 Mydict구현, 같은 key갑을 가지는 사전객체의 value값을 더한다.
class MyDict(dict):
    def __add__(self,value):
        for k in value:
            if k in self:
                self[k] += value[k]
            else :
                self[k] = value[k]
        return self

def cmp_1(a1, a2):
    return cmp(a2[1], a1[1])

class WebWordsFrequency:
    #객체생성과 함께 입력받은 url에 대하여 url과 단어빈도과 저장된 사전을 key와 value로 self.dict에 저장한다.
    def __init__(self,*args):
        self.dict = {}
        for url in args:
            self.dict[url] = WebWordsFrequency.arrangeURL(url)
    #url을 사전객체에 {'url':wordsFrequency사전객체}형식으로 저장
    def addURL(self,url):
        self.dict[url] = WebWordsFrequency.arrangeURL(url)
    # url을 self.dict에서 제거
    def removeURL(self,url):
        del self.dict[url]
    # self.dict의 key 출력
    def listURL(self):
        for url in self.dict:
            print url
    # + 연산을 하도록 self.dict를 잠시 MyDict로 형변환한 후에 비어있는 MyDict객체 d에 다 더해서 종합단어장을 만든다. 
    def getWordsFrequency(self):
        d = MyDict({})
        for key in self.dict:
            d += MyDict(self.dict[key])
        return d
    # getWordsFrequency에서 종합단어장을 받아와서 튜플형 리스트로 변환한 후에 두 번재 요소(value)값으로 비교해서 정렬한다.
    def getMaxFrequency(self):
        if self.dict == {}:#인수로 받은 url이 없으면 None반환
            return None
        d = self.getWordsFrequency()
        t = d.items()
        t.sort(cmp_1)
        l = []
        for tup in t:#제일 큰 value를 갖는 단어와 같은 value를 갖는 단어가 있으면 같이 반환한다.
            if tup[1] == t[0][1]:
                l.append(tup[0])
        return l
    #입력된 word와 가장 높은 유사도를 가지는 url을 출력한다.
    def searchURLByWord(self,word):
        similarity = {}
        for s_url in self.dict:# 각 url별로 유사도를 저장할 새 사전을 만듭니다. 
            similarity[s_url] = 0
        for url in similarity:
            d = self.dict[url]#해당 url의 단어장을 얻어옵니다.
            for key in d:#단어마다 word랑 포함관계가 있는지 검사하고, 있으면 빈도 값을 similarity에 추가합니다.
                 if key.lower() in word.lower() or word.lower() in key.lower():
                        similarity[url] += d[key]
        best = similarity.items()
        best.sort()
        l = []
        for url in best:#제일 큰 유사도를 갖는 url와 같은 유사도를 갖는 url이있으면 같이 반환한다.
            if url[1] == best[0][1]:
                l.append(url[0])
        for i in l:
            print i         
    @staticmethod
    def removeStopWords(lst):#영어 불용어를 제거 합니다.
        lst2 = []
        for i in range(len(lst)):
            if lst[i] in stopwords:
                continue
            lst2.append(lst[i])
        return lst2
    @staticmethod
    def arrangeURL(url):
        source = urllib2.urlopen(url).read()
        s1 = ''
        # 추후 숙제 답안 제출 때엔 왼쪽 라인은 주석처리할 것
        boolean = True #flag 변수
        for i in source:
            if i == '<':
                boolean = False
                continue
            if i == '>':
                boolean = True
                continue
            # '<'가 나왔을 때 '>'가 나올 때까지 아무연산도 하지않음
            if boolean:
                if i in string.punctuation:
                    i = ' '
                s1 += i
            # " < > "밖에 있는 문자를 구두문자를 제거해서 모두 s1에 저장
        t =  s1.split()# s1의 공백정리 및 리스트화
        t = WebWordsFrequency.removeStopWords(t)#영어 불용어 제거
        d = {}
        for x in t:             #사전 d에 리스트 t의 요소 key로 추가하고 값을 1로 합니다. 
            if x in d:          #다만 중복이 발생하면 key를 추가하지 않고, 이미 있는 key의 값을 1추가합니다.
                d[x] += 1
            else:
                d[x] = 1
        return d
w1 = WebWordsFrequency("http://www.nytimes.com/","http://www.cloudera.com/","http://www.bmw.com/")
w1.listURL()
print 
w1.addURL("http://marvel.com/")
w1.listURL()
print
w1.removeURL("http://www.bmw.com/")
w1.listURL()
print
#print w1.getWordsFrequency()
print
print w1.getMaxFrequency()
print
w1.searchURLByWord('doctor')
```

### 8번. 7번 문제에서 정의한 WebWordsFrequency 클래스를 상속하여 OrderedWebWordsFrequency 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드 하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def __init__(self,*args):
        WebWordsFrequency.__init__(self,*args)
        self.Iter = sorted(WebWordsFrequency.getWordsFrequency(self).items(),cmp_1)
        #종합단어장을 빈도수로 내림차순으로 만든 리스트를 객체의 변수에 저장합니다.
    def getWordsFrequency(self,reverse = False):
        d = WebWordsFrequency.getWordsFrequency(self)#부모클래스의 함수를 사용합니다.
        lst = d.items()
        if reverse == False:#내림차순
            lst.sort(cmp_1)
            print lst
        else:#올림차순
            lst.sort(cmp_1,reverse = True)
            print lst
    def __iter__(self):#next()함수를 가진 객체를 반환합니다.
        return self
    def next(self):
        if self.Iter == []:#단어장이 비면 다시채워주고, 반복을 끝냅니다.
            self.Iter = sorted(WebWordsFrequency.getWordsFrequency(self).items(),cmp_1)
            raise StopIteration
        result = self.Iter.pop(0)#단어장의 앞에서 부터 순서대로 하나씩 반환합니다. 
        return result
```
